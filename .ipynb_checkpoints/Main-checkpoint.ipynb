```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 
  
# Importing the dataset 
df = pd.read_excel('data.xlsx') 
df 
```

```python
import datetime as dt
df['Date'] = pd.to_datetime(df['Column1'])
df['Date'] = df['Date'].map(dt.datetime.toordinal)
df
```

```python
df['Date_2'] = df['Date'].map(dt.datetime.fromordinal)
df
```

```python
x = df.iloc[:,6:7].values
print(x)
y = df.iloc[:,1].values
print(y)
```

```python
from sklearn.preprocessing import PolynomialFeatures 
from sklearn.linear_model import LinearRegression 
 
poly = PolynomialFeatures(degree = 4) 
x_fit = poly.fit_transform(x) 
  
poly.fit(x_fit, y) 

lin2 = LinearRegression() 
lin2.fit(x_fit, y) 
```

```python
plt.scatter(x, y, color = 'blue') 
  
plt.plot(x, lin2.predict(poly.fit_transform(x)), color = 'red') 
plt.title('Polynomial Regression') 
plt.xlabel('Cases') 
plt.ylabel('date') 
  
plt.show() 
```

```python
predict = np.array([i for i in range(737497,737600)])
print(predict)
```

```python
pr = predict.reshape(-1,1)
pr_real = [dt.datetime.fromordinal(i) for i in pr]

predictions = lin2.predict(poly.fit_transform(pr)) 
predictions
```

```python

fig, ax = plt.subplots()


ax.scatter(df['Column1'], y, color = 'blue', s=5) 
ax.scatter(pr_real, predictions, color = 'red', s=5) 

  
plt.plot(df['Column1'], lin2.predict(poly.fit_transform(x)), color = 'green') 
plt.plot(pr_real, predictions, color = 'green')

plt.title('Polynomial Regression') 
plt.xlabel('Dates') 
plt.ylabel('Cases') 

plt.savefig('foo.pdf')


plt.show()


```
